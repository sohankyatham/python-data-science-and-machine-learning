## Matplotlib
- Used to create plots and charts such as Line plots, Bar charts, Scatter plots, Histograms, & Pie charts

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Line Plot")

plt.plot(x, y, color='blue', linestyle='--', marker='o')

plt.show()

'''
plt.plot(x, y) → draws a line

plt.show() → displays the plot

Colors: 'red', 'blue', 'green'

Linestyles: '-', '--', ':'

Markers: 'o', 's', '^'
'''
```

```python
x = [1, 2, 3, 4]

plt.plot(x, [10, 20, 30, 40], label="Line 1")
plt.plot(x, [5, 15, 25, 35], label="Line 2")

plt.legend()
plt.show()
```

```python
# Save a plot
plt.plot(x, y)
# plt.savefig("my_plot.png")
plt.show()
```

```python
# Bar Chart
names = ["A", "B", "C"]
values = [5, 7, 3]

plt.bar(names, values)
plt.title("Bar Chart Example")
plt.show()
```

```python
# Scatter Plot
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    "height": [170, 180, 175],
    "weight": [65, 80, 70]
})

plt.scatter(df["height"], df["weight"])
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```
