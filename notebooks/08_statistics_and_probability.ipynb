# Probability & Statistics

## Central Limit Theorem
- If you take many large, random samples from any population, the averages (means) of those samples will form a bell-shaped (normal) distribution, even if the original population's data isn't normally distributed (e.g., skewed)
- The sampling distribution of the sample means approaches a normal distribution as the sample size gets larger
- Independent of distribution of the population

## Variance vs Covariance
- Variance measures the variation of a single random variable
- Covariance is a measure of how much two random variables vary together
