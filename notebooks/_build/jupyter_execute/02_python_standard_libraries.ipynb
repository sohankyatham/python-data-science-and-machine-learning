```python
# random - used for simulations and sampling
import random

random.random()
random.randint(1, 10)

sequence = ["apple", "banana", "cherry", "date"]
selected_item = random.choice(sequence)
print("This is the randomly selected item:", selected_item)

shuffled_list = random.shuffle(sequence)
print("This is the shuffled list:", sequence)
```
