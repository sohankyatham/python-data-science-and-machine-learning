## Pandas 

- A package built on top of NumPy

```python
# DataFrame
import pandas as pd

df = pd.DataFrame({
    "height": [170, 180, 175],
    "weight": [65, 80, 70]
})

print("This is the data:")
print(df)
```

```python
# Accessing Data
print(df["height"])
print(df.loc[0])
print(df.loc[0, "height"])
```

```python
# Filtering Rows
df[df["height"] > 175]

# Creating New Columns
df["bmi"] = df["weight"] / (df["height"]/100)**2
```
